```python
def add(a,b):
    return(a+b)
    
def sub(a,b):
    return(a-b)
    
    
def mul(a,b):
    return(a*b)
    
    
    
def div(a,b):
    return(a/b)    


def cal():
    

    a = int(input("Enter A = "))
    b = int(input("Enter B = "))
    o = input("Enter + - * / ")
    if o == "+":
        print(add(a,b))
    
    elif o == "-":
        print(sub(a,b))
    
    elif o == "*":
        print(mul(a,b))

    elif o == "/":
        print(div(a,b))
    
    else:
        print("Invalid operator")
        
    z = input('Enter y for more calculation')
    if z == 'y':
        cal()
cal()
```

```python
def cal(l,b,h):
    a = l*b
    v = a*h
    return a,v

l = int(input("Enter l = "))
b = int(input("Enter b = "))
h = int(input("Enter h = "))    
area,volume = cal(l,b,h)

print('The area is = ', area)
print('The volume is = ', volume)
```
