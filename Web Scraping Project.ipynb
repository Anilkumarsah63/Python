```python
tags = (data.table).find_all('tr')
tags
```

```python
list_data = []

for i in tags:
    x = i.get_text().split()
    list_data.append(x)
    
list_data    
```

```python
import csv
with open('covid.csv','a')as f:
    x = csv.writer(f)
    for i in final_list:
        x.writerow(i)
        
```
